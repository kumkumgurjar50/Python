```python
# write a python program using Beautifulsoup library create dataframe of
#statement, link, source, date 
#from "https://www.politifact.com/"
import requests
from bs4 import BeautifulSoup
response=requests.get("https://www.politifact.com/")
response=response.text
soup=BeautifulSoup(response,"html.parser")
```

```python
source=[]
statement=[]
date=[]
link=[]
a=soup.find("div",class_="embla")
#print(a)
b=a.find_all("div",class_="embla__slide caro-listing__item")
#print(b)
for i in b:
    s=i.find("div",class_="m-statement__meta").find("a").text.strip()
    source.append(s)
    statement.append(i.find("div",class_="m-statement__quote").find("a").text.strip())
    link.append("https://www.politifact.com"+i.find("div",class_="m-statement__quote").
        find("a")["href"])
    d=i.find("div",class_="m-statement__desc").text.strip()
    l=d.split()
    date.append((" ".join(l[2:5])))
import pandas as pd
df=pd.DataFrame({"source":source,"statement":statement,"link":link,"date":date})
print(df)
```

```python
# write a python program to scrap data from "30 Most Popular Movies Right Now_ What to Watch In Theaters and Streaming __ 
#Rotten Tomatoes – Movie and TV News" and create dataframe with columns movie 
#name, rating, director
```

```python
from bs4 import BeautifulSoup
file=open("30 Most Popular Movies Right Now_ What to Watch In Theaters and Streaming __ Rotten Tomatoes – Movie and TV News.html",
          encoding="utf-8")
response=file.read()
soup=BeautifulSoup(response,"html.parser")
movie=[]
rating=[]
director=[]
a=soup.find_all("div",class_="col-sm-18 col-full-xs countdown-item-content")
for i in a:
    movie.append(i.find("h2").find("a").text.strip())
    rating.append(i.find("span",class_="tMeterScore").text.strip())
    director.append(i.find("div",class_="info director").find("a").text.strip())

import pandas as pd
df=pd.DataFrame({"movie":movie,"rating":rating,"director":director})
df
```

```python
# write a python program to scrap data from "Certified used Mercedes-Benz for sale.html"
#create dataframe with columns carname, review count, rating, price,dealer name, mileage
```

```python
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
file=open("Certified used Mercedes-Benz for sale.html",
          encoding="utf-8")
response=file.read()
soup=BeautifulSoup(response,"html.parser")
carname=[]
review_count=[] 
rating=[]
price=[]
dealer=[]
name=[]
mileage=[]
a=soup.find_all("div",class_="vehicle-details")
for i in a:
    carname.append(i.find("h2",class_="title").text.strip())
    mileage.append(i.find("div",class_="mileage").text.strip())
    review_count.append(i.find("span",class_="sds-rating__link sds-button-link").
        text.strip().split()[0][1:])
    dealer.append(i.find("div",class_="dealer-name").text.strip())
    price.append(i.find("span",class_="primary-price").text.strip())
    try:
        rating.append(i.find("span",class_="sds-rating__count").text.strip())
    except:
        rating.append(np.nan)
        
df=pd.DataFrame({"car name":carname,"mileage":mileage,"review_count":review_count,
               "dealer":dealer,"price":price,"rating":rating })
df
```

```python
# Pagination
import requests
url = "https://jsonplaceholder.typicode.com/posts"
params = {
"_page": 1, # no of page
"_limit": 5 # no of post per page
}
response = requests.get(url, params=params)
data = response.json()
print("Posts on page 1")
for post in data:
    print(post["id"], post["title"])
```

```python
## Multiple pages
import requests
base_url = "https://jsonplaceholder.typicode.com/posts"
page = 1
while page <= 3:
    params = {
    "_page": page,
    "_limit": 5
    }
    response = requests.get(base_url, params=params)
    posts = response.json()
    print(f"\nPage {page}")
    for post in posts:
        print(post["id"], post["title"])
    page += 1
```
