```python
import matplotlib.pyplot as plt
import numpy as np
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])
plt.plot(y1)
plt.plot(y2)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
plt.xlabel("X-axis",fontdict=font1)
plt.ylabel("Y-axis",fontdict=font2)
plt.title("Title",fontdict=font2,loc="left")
plt.grid(axis="x")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,1,2,3])
y=np.array([3,8,1,0])
plt.subplot(2,3,1)
plt.plot(x,y)
plt.title("Title",color="r")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

x=np.array([0,1,2,3])
y=np.array([3,8,1,0])
plt.subplot(2,3,2)
plt.plot(x,y)
plt.title("Title",color="r")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")

x=np.array([0,1,2,3])
y=np.array([3,8,1,0])
plt.subplot(2,3,3)
plt.scatter(x,y)
plt.title("Title",color="r")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.colorbar()
plt.suptitle("Sup Title")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=[10,20,30,40,50]
l=["A","B","C","D","E"]
e=[0.1,0,0,0,0]
plt.pie(x,labels=l,explode=e,shadow=True)
plt.legend()
plt.show()
```
